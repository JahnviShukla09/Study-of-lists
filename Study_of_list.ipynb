```python
#Insert items
thislist = ["orange", "grapes", "kiwi"]
thislist.insert(2,"apple")
print(thislist)
```

```python
#Extend list
thislist = ["orange", "grapes", "kiwi"]
mylist = ["apple", "banana", "chery"]
thislist.extend(mylist)
print(thislist)
```

```python
numbers = [12, 15, 17, 20, 25]
print("Elements from index 1 to 3", numbers[1:4])
print("First three elements", numbers[0:3])
print("Elements from index 2 onwards(Reverse indexing)", numbers[-3:])
print("All elments", numbers[:])
```

```python
marks = [67, 78, 59, 75, 70]
print("Maximum marks are:", max(marks))
print("Minimum marks are:", min(marks))
print("Total marks are:", sum(marks))
print("Number of students:", len(marks))
print("Average marks", [sum(marks)/len(marks)])
marks.sort()
print(marks)
```

```python
list = ["soap", "toothpaste", "fruits", "vegetables", "snacks", "flour"]
print("Grocery list:", list)
list.insert(2,"dishwash")
print("Insert an item: ", list)
list.remove("snacks")
print("Remove an item: ", list)
print("Final list:", list)
```

```python
list = ["jahnvi", "falak", "hiranya", "divisha"]
print("List of contcts:", list)
list.insert(4,"rivya")
print("Add a new contact: ", list)
list.remove("divisha")
print("Delete a contact: ",list)
list.sort()
print("Contacts alphabetically: ", list)
```

```python
list1 = [20,30,50,60,40,10,70]
print(list1[1:3])
print(list1[1:3:1])
print(list1[1:5:2])
print(list1[1:4:3])
print(list1[6:4:-1])
print(list1[-2:-4:-2])
```

```python
#Temperature analysis
temp = [24, 23, 26, 21, 27, 25, 22, 20, 29, 28]
print("The temperature of first 5 days:", temp[0:5])
print("The temperature of last 5 days:", temp[-5:])
print("Highest temperature:", max(temp))
print("Lowest temperature:", min(temp))
print("Average temperature:", [sum(temp)/len(temp)])
```

```python
#Daily expense tracker
expense = [100, 250, 150, 275, 300, 200, 320]
print("The total expense is:", len(expense))
print("The average expense is:", [sum(expense)/len(expense)])
print("The highest spending day is:", max(expense))
print("The lowest spending day is:", min(expense))
```
